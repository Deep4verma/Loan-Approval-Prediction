```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
```

```python
df=pd.read_csv("bankloan.csv",index_col="ID")
df.head()
```

```python
df.shape
```

```python
df_personal_loan=df["Personal.Loan"]
df=df.drop(["Personal.Loan"],axis=1)
df["Personal.Loan"]=df_personal_loan
df.head()
```

```python
(df["Personal.Loan"]==0).sum()
```

```python
df1=df[df["Personal.Loan"]==0]
df2=df[df["Personal.Loan"]==1]
df1
```

```python
df1=df.sample(n=1000,random_state=10)
```

```python
df_final=pd.concat([df1,df2],axis=0)
df_final.describe()
```

```python
corre_=df_final.corr()
plt.figure(figsize=(12,10))
sns.heatmap(corre_,annot=True)
plt.show()
```

univariate analysis

```python
df_final=df_final.drop(["Securities.Account","Online","CreditCard","ZIP.Code"],axis=1)
X=df_final.drop(["Personal.Loan"],axis=1)
Y=df_final["Personal.Loan"]
Y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=12)
```

```python
plt.figure(figsize=(12,5))
plt.subplot(121)
sns.countplot(x="Personal.Loan", data=df_final)
plt.subplot(122)
sns.distplot(df_final["Personal.Loan"],kde=True)
plt.show()
```

```python
import scipy.stats as stats
```

```python
plt.figure(figsize=(24,5))
plt.subplot(131)
sns.boxplot(df_final["Income"])
plt.subplot(132)
sns.distplot(df_final["Income"])
plt.subplot(133)
stats.probplot(df_final["Income"],dist="norm",plot=plt)
plt.xlabel("income")
plt.ylabel("frequancy")
plt.show()
```

```python
plt.figure(figsize=(24,5))
plt.subplot(132)
sns.boxplot(df_final["CCAvg"])
plt.subplot(131)
sns.distplot(df_final["CCAvg"],hist=False,kde=True)
plt.subplot(133)
stats.probplot(df_final["CCAvg"],dist="norm",plot=plt)
plt.xlabel("CCAvg")
plt.ylabel("frequancy")
plt.show()
```
