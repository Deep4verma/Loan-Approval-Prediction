```python
def outliers_details(df,df_num_col):
  lower_lim_list=[]
  upper_lim_list=[]
  outlier_num=[]
  outlier_pct=[]
  for i in df_num_col:
    Q1=df[i].quantile(0.25)
    Q3=df[i].quantile(0.75)
    IQR=Q3-Q1
    lower_lim=round((Q1-1.5*IQR),0)
    upper_lim=round((Q3+1.5*IQR),0)
    lower_lim_list.append(lower_lim)
    upper_lim_list.append(upper_lim)
    outliers_in_col=df[(df[i]<lower_lim)|(df[i]>upper_lim)]
    outlier_num.append(len(outliers_in_col))
    outlier_pct.append(round(len(outliers_in_col)/len(df),3)*100)
  outlier_def=pd.DataFrame({"column":df_num_col,"lower_limit":lower_lim_list,"upper_limit":upper_lim_list,
                              "number_of_outliers":outlier_num,"percentage_of_outlier":outlier_pct})
  return outlier_def
col=["Age","CCAvg","Income","Mortgage"]
outliers_details(df,col)
```

here i saw that


*   **ccavg have outliers less that 5 percent that means i can removie these outliers**
* *mortagage have very high outlier percentage   



```python
for i in col[0:2]:
  Q1=df[i].quantile(0.25)
  Q3=df[i].quantile(0.75)
  IQR=Q3-Q1
  lower_lim=Q1-IQR*1.5
  uppr_lim=Q3+1.5*IQR
  df_filt=df[(df[i]>=lower_lim)&(df[i]<=uppr_lim)]
outliers_details(df_filt,col)
```

here column mortgage have outllier more tha 5 percent so we do function transform to adjust the shape and scale

```python
df_filt.shape
```

```python

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import FunctionTransformer,StandardScaler
X=df_filt.drop(["Personal.Loan"],axis=1)
Y=df_filt["Personal.Loan"]
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.3,random_state=10)
```

```python
trf=FunctionTransformer(func=np.log1p)
x_train_trans=x_train
x_test_trans=x_test
x_train_trans["Mortgage"]=round(trf.fit_transform(x_train["Mortgage"]),2)
x_test_trans["Mortgage"]=round(trf.transform(x_test["Mortgage"]),2)
x_train_trans.head()
data_final=pd.DataFrame(x_train_trans,columns=x_train_trans.columns).reset_index(drop=True)
outliers_details(data_final,col)
```

```python
scaler=StandardScaler()
x_train_scaled=scaler.fit_transform(x_train_trans)
x_test_scaled=scaler.transform(x_test_trans)
x_train_scaled=np.round(x_train_scaled,2)
x_test_scaled=np.round(x_test_scaled,2)
data_1=pd.DataFrame(x_train_scaled,columns=x_train.columns)
data_1.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
para_ran={"n_estimators":[10,20,50,100],"max_depth":[10,20,50,100,200]}
grid=GridSearchCV(RandomForestClassifier(random_state=10),para_ran,cv=10)
grid.fit(x_train_scaled,y_train)
best_para=grid.best_params_
best_score=round(grid.best_score_,2)
print(f"the best parameters are {best_para}")
print(f"the best score is {best_score}")
```

```python
crf=RandomForestClassifier(max_depth=10,n_estimators=100,random_state=42)
crf.fit(x_train_scaled,y_train)
y_pred=crf.predict(x_test_scaled)
score=crf.score(x_test_scaled,y_test)
print(f"accuracy of the model is {score*100}")
```
