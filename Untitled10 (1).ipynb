```python
df=df[df["Experience"]>=0]
df.shape
```

```python
df.describe()
```

```python
Q1=df["CCAvg"].quantile(0.25)
Q3=df["CCAvg"].quantile(0.75)
IQR=Q3-Q1
lower_lim=Q1-1.5*IQR
upper_lim=Q3+1.5*IQR
df_final=df[(df["CCAvg"]>lower_lim)&(df["CCAvg"]<upper_lim)]
df_final.shape
```

```python
df_final.describe()
```

```python
Q1=df["Mortgage"].quantile(0.25)
Q3=df["Mortgage"].quantile(0.75)
IQR=Q3-Q1
lower_lim=Q1-1.5*IQR
upper_lim=Q3+1.5*IQR
df_final=df[(df["Mortgage"]>lower_lim)&(df["Mortgage"]<upper_lim)]
df_final.shape
```

```python
df_final.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
X=df_final.drop(["Personal.Loan"],axis=1)
Y=df_final["Personal.Loan"]
```

```python
para_ran={"n_estimators":[10,20,50,100],"max_depth":[2,50,100,200]}
grid= GridSearchCV(RandomForestClassifier(),para_ran,cv=10)
grid.fit(X,Y)
```

```python
best_score=round(grid.best_score_,2)
best_score
```

```python
grid.best_estimator_
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.3,random_state=10)
model1=RandomForestClassifier(n_estimators=50,max_depth=50)
model1.fit(x_train,y_train)
```

```python
model1.score(x_test,y_test)
```
